**14. Convertors**

```python
def rename(name):
  if name=='Royal Challengers Bangalore':
    return 'RCB'
  else:
    return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename}).head()
```
